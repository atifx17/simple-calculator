```python
from tkinter import*
root=Tk()
root.geometry("375x320")
root.resizable(0,0)
root.title("Calculator")
a=StringVar()
e=Entry(root,borderwidth=5,font=("arial",20),justify="right",textvariable=a)
e.place(height=50,width=375)

def click(p):
    x=a.get()
    if len(x)==1 and x[0]=="0":
        a.set(p)
    else:
        a.set(x+p)
def op(o):
    x=a.get()
    a.set(x+o)

def clear():
    a.set("0")


def equal():
    x=a.get()
    if ("+" in x):
        l=list(map(float,x.split("+")))
        sum=0
        for i in l:
            sum+=i
        a.set(sum)
    elif ("-" in x):
        l=list(map(float,x.split("-")))
        a.set(l[0]-l[1])
    elif ("/" in x):
        l=list(map(float,x.split("/")))
        a.set(l[0]/l[1])
    elif ("*" in x):
        l=list(map(float,x.split("*")))
        a.set(l[0]*l[1])
    elif ("√" in x):
        l=x.split("√")
        a.set(float(l[1])**0.5)
    elif ("!" in x):
        l=x.split("!")
        n=int(l[0])
            
        if n < 0:
            pass
        elif n == 0 or n == 1:
            result=1
        else:
            result = 1
            for i in range(2, n + 1):
                result *= i
            
        a.set(result)
    
    
    elif ("^3" in x):
        l=x.split("^3")
        a.set(float(l[0])**3)
    else:
        a.set("error")


#buttons
c=Button(root,text="C",padx=30,pady=10,command=clear)
eq=Button(root,text="=",padx=30,pady=10,command=equal)
rq=Button(root,text="√",padx=30,pady=10,command=lambda:op("√"))
div=Button(root,text="/",padx=30,pady=10,command=lambda:op("/"))
mul=Button(root,text="*",padx=30,pady=10,command=lambda:op("*"))
sub=Button(root,text="-",padx=30,pady=10,command=lambda:op("-"))
add=Button(root,text="+",padx=30,pady=10,command=lambda:op("+"))
sq=Button(root,text="!",padx=30,pady=10,command=lambda:op("!"))
cb=Button(root,text="^2",padx=30,pady=10,command=lambda:op("^2"))

b7=Button(root,text="7",padx=30,pady=10,command=lambda:click("7"))
b8=Button(root,text="8",padx=30,pady=10,command=lambda:click("8"))
b9=Button(root,text="9",padx=30,pady=10,command=lambda:click("9"))
b4=Button(root,text="4",padx=30,pady=10,command=lambda:click("4"))
b5=Button(root,text="5",padx=30,pady=10,command=lambda:click("5"))
b6=Button(root,text="6",padx=30,pady=10,command=lambda:click("6"))
b1=Button(root,text="1",padx=30,pady=10,command=lambda:click("1"))
b2=Button(root,text="2",padx=30,pady=10,command=lambda:click("2"))
b3=Button(root,text="3",padx=30,pady=10,command=lambda:click("3"))
b0=Button(root,text="0",padx=30,pady=10,command=lambda:click("0"))
dot=Button(root,text=".",padx=30,pady=10,command=lambda:click("."))
c.place(x=0,y=55)
rq.place(x=100,y=55)
div.place(x=200,y=55)
mul.place(x=300,y=55)

b7.place(x=0,y=110)
b8.place(x=100,y=110)
b9.place(x=200,y=110)
sub.place(x=300,y=110)

b4.place(x=0,y=165)
b5.place(x=100,y=165)
b6.place(x=200,y=165)
add.place(x=300,y=165)

b1.place(x=0,y=220)
b2.place(x=100,y=220)
b3.place(x=200,y=220)
eq.place(x=300,y=220)

dot.place(x=0,y=275)
b0.place(x=100,y=275)
sq.place(x=200,y=275)
cb.place(x=300,y=275)

root.mainloop()
```
